```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_gray=cv2.imread("./ztyo.jpg",0)
img_eq_hist=cv2.equalizeHist(img_gray)
fig=plt.figure(figsize=(10,7))
rows=1
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(img_gray,cmap="gray")
plt.axis('off')
plt.title("BÄ±nary threshholding")
fig.add_subplot(rows,columns,2)
plt.imshow(img_eq_hist,cmap="gray")
plt.axis('off')
plt.title("BÄ±nary threshholding")
            
```

```python
my_original_img=cv2.imread("./ztyo.jpg",1)
```

```python
my_kernel=np.ones((5,5),dtype=np.uint8)
my_kernel
```

```python
my_erosion_img=cv2.erode(my_original_img,my_kernel,iterations=2)
fig=plt.figure(figsize=(10,7))
rows=1
columns=2
fig.add_subplot=(rows,columns,1)
plt.imshow(my_original_img,cmap="gray")
plt.axis('off')
plt.title("OrÄ±gÄ±nal Image")

fig.add_subplot=(rows,columns,2)
plt.imshow(my_erosion_img,cmap="gray")
plt.axis('off')
plt.title("Erosion Image")
```

```python
fig= plt.figure(figsize=(10 ,7))
rows =1
columns =3
fig.add_subplot(rows ,columns ,1)
plt.imshow(my_original_img , cmap="gray")
plt.axis("off")
plt.title("Original Image")

fig.add_subplot(rows,columns ,2)
plt.imshow(my_erosion_img , cmap="gray")
plt.axis("off")
plt.title("EROSIONÂ Image")
```

```python
my_dilation=cv2.dilate(my_original_img,my_kernel,iterations=2)
fig=plt.figure(figsize=(10,7))
rows=1
columns=3
fig.add_subplot(rows ,columns ,1)
plt.imshow(my_original_img , cmap="gray")
plt.axis("off")
plt.title("Original Image")
fig.add_subplot(rows ,columns ,2)
plt.imshow(my_erosion_img , cmap="gray")
plt.axis("off")
plt.title("Erosion Image")
fig.add_subplot(rows ,columns ,3)
plt.imshow(my_dilation , cmap="gray")
plt.axis("off")
plt.title("DilatÄ±on Image")
```

```python
new_img=cv2.imread("./ztyobst.jpg",0)
white_noise=np.random.randint(0,2,size=new_img.shape[:2])
white_noise=white_noise*255
my_dilation=cv2.dilate(my_original_img,my_kernel,iterations=2)
fig=plt.figure(figsize=(10,7))
rows=1
columns=3
fig.add_subplot(rows ,columns ,1)
plt.imshow(white_noise , cmap="gray")
plt.axis("off")
plt.title("Original Image")
```

```python
my_noised_img=new_img+white_noise
plt.imshow(my_noised_img,cmap="gray")
```

```python
my_opening_img=cv2.morphologyEx(my_noised_img.astype(np.float32),
                               cv2.MORPH_OPEN,my_kernel)
```

```python
fig=plt.figure(figsize=(10,7))
rows=1
columns=3
fig.add_subplot(rows ,columns ,1)
plt.imshow(new_img , cmap="gray")
plt.axis("off")
plt.title("Original Image")
fig.add_subplot(rows ,columns ,2)
plt.imshow(my_noised_img , cmap="gray")
plt.axis("off")
plt.title("Noised Image")
fig.add_subplot(rows ,columns ,3)
plt.imshow(my_opening_img, cmap="gray")
plt.axis("off")
plt.title("Opening  Image")
```

```python
black_noise=np.random.randint(0,2,size=new_img.shape[:2])
black_noise=black_noise*-255
```

```python
print(np.max(black_noise))
print(np.min(black_noise))
```

```python
plt.imshow(black_noise,cmap="gray")
```

```python
my_black_noised_img=new_img+black_noise
plt.imshow(my_black_noised_img,cmap="gray")
```

```python
my_black_noised_img[my_black_noised_img<-245]=0
```

```python
my_closing_img=cv2.morphologyEx(my_noised_img.astype(np.float32),
                               cv2.MORPH_CLOSE,my_kernel)
```

```python
fig=plt.figure(figsize=(10,7))
rows=1
columns=3
fig.add_subplot(rows ,columns ,1)
plt.imshow(new_img , cmap="gray")
plt.axis("off")
plt.title("Original Image")
fig.add_subplot(rows ,columns ,2)
plt.imshow(my_black_noised_img , cmap="gray")
plt.axis("off")
plt.title("Black Image")
fig.add_subplot(rows ,columns ,3)
plt.imshow(my_closing_img, cmap="gray")
plt.axis("off")
plt.title("Closing  Image")
```

```python
my_gradient_img_1=cv2.morphologyEx(my_original_img,cv2.MORPH_GRADIENT,my_kernel)
my_gradient_img_2=cv2.morphologyEx(new_img,cv2.MORPH_GRADIENT,my_kernel)
```

```python
fig=plt.figure(figsize=(10,7))
rows=2
columns=2
fig.add_subplot(rows ,columns ,1)
plt.imshow(my_original_img , cmap="gray")
plt.axis("off")
plt.title("Original Image 1")
fig.add_subplot(rows ,columns ,2)
plt.imshow(my_gradient_img_1 , cmap="gray")
plt.axis("off")
plt.title("Gradiant Image 1")
fig.add_subplot(rows ,columns ,3)
plt.imshow(new_img, cmap="gray")
plt.axis("off")
plt.title("OrigÄ±nal  Image 2")
fig.add_subplot(rows ,columns ,4)
plt.imshow(my_gradient_img_2, cmap="gray")
plt.axis("off")
plt.title("Gradiant  Image 2")
```
